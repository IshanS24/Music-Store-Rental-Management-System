```python
### musicReturn.py


from datetime import datetime, date

# Load Subscription, Music, and Rental data
Sub_info = []
with open("Subscription_Info.txt", "r") as f:
    f.readline()  # Skip the first line
    for line in f:
        s = line.strip()
        Sub_info.append(list(s.split(",")))

Music_info = []
with open("music_Info.txt", "r") as m:
    m.readline()  # Skip the first line
    for line in m:
        s = line.strip()
        Music_info.append(list(s.split("|")))

Rental_info = []
with open("Rental.txt", "r") as r:
    r.readline()  # Skip the first line
    for line in r:
        s = line.strip()
        Rental_info.append(list(s.split(",")))


# Validate active subscription
def activeSubscription():
    subID = str(input("Enter subscription ID: "))
    for record in Sub_info:
        if record[0] == subID:
            date_string = record[3]
            try:
                parsed_date = datetime.strptime(date_string, "%Y-%m-%d").date()
                current_date = date.today()
                if parsed_date > current_date:
                    print("Subscription is active.")
                    return subID, record[1].strip()  # Return customer type
                else:
                    print("Subscription is expired.")
                    return None, None
            except ValueError:
                print(f"Invalid date format in subscription record: {date_string}")
                return None, None
    print("Subscription ID not found.")
    return None, None


# Main rent function
def rent():
    subID, customer_type = activeSubscription()
    if subID:
        rentID = input("ID of record you wish to rent? ")
        if not exists(rentID):
            print(f"ERROR: Record {rentID} does not exist in the music inventory.")
            return

        # Check if the customer has reached rental capacity
        rented = 0
        for rec in Rental_info:
            if len(rec) < 4:  # Check if the record is incomplete
                print(f"WARNING: Skipping incomplete record {rec}")
                continue  # Skip incomplete records

            if rec[3] == subID:  # Check if the customer ID matches
                rented += 1  # Increment the counter if it matches
                print(rented)
        print(customer_type)
        if customer_type == "Basic" and rented >= 2:
            print("Max rental capacity reached for Basic customer. Cannot rent more than 2 records.")
            return
        elif customer_type == "Premium" and rented >= 7:
            print("Max rental capacity reached for Premium customer. Cannot rent more than 7 records.")
            return
        
        # Proceed with rental check
        record_index = get_rental_index(rentID)
        if record_index is None:    # If the record is not already rented
            print(f"Record {rentID} is available for a new rental.")
            add_new_rental(rentID, subID)
        else:   # If the record is already rented, check its status
            isRented(record_index, rentID, subID, customer_type)



# Check if record exists in music_info
def exists(rentID):
    for rec in Music_info:
        if rec[0] == rentID:
            return True
    return False



# Get index of rental record or None
def get_rental_index(rentID):
    index = 0  # Initialize the index counter
    for rec in Rental_info:
        if rec[0] == rentID:
            return index
        index += 1  # Increment the index counter
    return None



# Handle rental process
def isRented(record, rentID, subID, customer_type):
    rental_date = Rental_info[record][1].strip()
    return_date = Rental_info[record][2].strip()

    if rental_date == "" and return_date == "":
        print(f"Record {rentID} is available for rent.")
        update_rental(record, subID)
    elif return_date:
        return_date_obj = datetime.strptime(return_date, "%d-%m-%Y").date()
        if return_date_obj <= date.today():
            print(f"Record {rentID} is available for rent (previously returned).")
            add_new_rental(rentID, subID)
        else:
            print(f"Record {rentID} is currently rented; return date not reached.")
    else:
        print(f"Record {rentID} is currently rented.")

# Add a new rental record
def add_new_rental(rentID, subID):
    current_date = datetime.now().strftime("%d-%m-%Y")
    new_rental = [rentID, current_date, "", subID]
    Rental_info.append(new_rental)
    update_music_info(rentID)
    write_rental_data()
    print(f"New rental created for record {rentID} by customer {subID}.")


# Update rental record
def update_rental(record, subID):
    current_date = datetime.now().strftime("%d-%m-%Y")
    Rental_info[record][1] = current_date
    Rental_info[record][2] = ""  # Clear return date
    Rental_info[record][3] = subID
    update_music_info(Rental_info[record][0])
    write_rental_data()
    print(f"Record {Rental_info[record][0]} rented successfully to customer {subID}.")


# Update Music_info purchaseID
def update_music_info(rentID):
    current_date = datetime.now().strftime("%d-%m-%y")  # Get the current date as Purchase Date
    for rec in Music_info:
        if rec[0] == rentID:  # Match on Record ID
            rec[5] = current_date  # change purchaseDate
            print(f"DEBUG: Updated Purchase Date for Record {rentID} to {current_date}.")
    write_music_data()  # Save changes back to file



# Write updated Rental data to file
def write_rental_data():
    with open("Rental.txt", "w") as rentalFile:
        rentalFile.write("Record ID,Rental Date,Return Date,Rental Customer ID\n")
        for rec in Rental_info:
            rentalFile.write(",".join(rec) + "\n")


# Write updated Music data to file
def write_music_data():
    with open("music_Info.txt", "w") as musicFile:
        musicFile.write("RecordID|Artist|Title|Medium|Genre|PurchaseDate\n")
        for rec in Music_info:
            musicFile.write("|".join(rec) + "\n")


# Start rental process
rent()

```
